```python
import pandas as pd
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
df = pd.DataFrame(data)

def get_wordnet_pos(treebank_tag):
    """Map 'treebank' part of speech tags to WordNet part of speech tags"""
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    pos_tags = nltk.pos_tag(tokens)
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(pos)) for token, pos in pos_tags]
    return ' '.join(lemmatized_tokens)

# Apply lemmatization to the 'headlines' column
df['lemmatized_headlines'] = df['headlines'].apply(lemmatize_text)

print(df[['headlines', 'lemmatized_headlines']])

```

```python
# Apply lemmatization to the 'description' column
df['lemmatized_description'] = df['description'].apply(lemmatize_text)

print(df[['description', 'lemmatized_description']])
```

```python
# Apply lemmatization to the 'content' column
df['lemmatized_content'] = df['content'].apply(lemmatize_text)

print(df[['content', 'lemmatized_content']])
```
